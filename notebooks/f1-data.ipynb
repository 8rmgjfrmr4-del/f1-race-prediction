```python
import requests
import pandas as pd

url = "https://api.jolpi.ca/ergast/f1/2023/results.json?limit=1000"
response = requests.get(url)

data = response.json()
data
```

```python
# Extract race results from JSON
results = data['MRData']['RaceTable']['Races']

# Normalize into tabular rows
df = pd.json_normalize(results, record_path=['Results'], 
                       meta=['raceName', 'round', 'season'])

df.head()
```

```python
df.to_csv("../data/f1_2023_results.csv", index=False)
```
